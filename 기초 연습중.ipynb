```python
class FourCal :
    def setdata(self, fir, sec) :  # first, second 객체를 생성해주는 함수
        self.first = fir  # first라는 객체 생성
        self.second = sec  # second라는 객체 생성
```

```python
a = FourCal()
```

```python
a.setdata(4,2)
```

```python
a.first
```

```python
a.second
```

```python
b = FourCal()
```

```python
b.setdata(1,7)
```

```python
b.first  # a에서의 first와는 다른 객체
```

```python
import pandas as pd
```

```python
obj = {'Score':[100, 95, 60, 77, 83, 85, 90, 90, 88, 75, 90, 54, 48, 84, 73], 
       'Age':[22, 22, 27, 18, 18, 19, 24, 26, 30, 27, 25, 21, 20, 17, 20]
      }
df = pd.DataFrame(obj)
df
```

```python
df.sort_values('Score', ascending = False, inplace = True)
df.head(5)
```

```python
df['rank_avg1'] = df['Score'].rank()
df['rank_avg2'] = df['Score'].rank(ascending = False)
df['rank_min'] = df['Score'].rank(method = 'min', ascending = False)
df['rank_max'] = df['Score'].rank(method = 'max', ascending = False)
```

```python
df
```
