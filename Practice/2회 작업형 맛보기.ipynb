```python
import pandas as pd

mtcars = pd.read_csv('./data/기출복원/02회/mtcars2.csv')
mtcars

mpg = mtcars['mpg'].copy()
drat = mtcars['drat'].copy()

tenth = mpg.sort_values(ascending = False, ignore_index = True)[9]
idx = mpg.sort_values(ascending = False)[:10].index

mpg[idx] = tenth

avg_mpg = mpg[drat >= 4].mean()
result = round(avg_mpg, 3)
print(result)
```

```python
import pandas as pd
mtcars = pd.read_csv('data/기출복원/02회/mtcars2.csv')

train_rate = mtcars.shape[0]*0.8
train = mtcars[:int(train_rate)]

disp_before = train['disp'].copy()
disp_after = train['disp'].copy()

med = disp_before.median()

disp_after.fillna(med, inplace = True)

std_before = disp_before.std()
std_after = disp_after.std()
result = round(abs(std_before - std_after), 3)

print(result)
```

#### 3. gehan 데이터 셋을 불러와 time 컬럼에서 이상값의 합을 구하여라.(단, 이상값은 평균에서 1.5 표준편차 이상으로 벗어난 값으로 정의함)

```python
import pandas as pd

gehan = pd.read_csv('data/기출복원/02회/gehan.csv')

time = gehan['time'].copy()

mean = time.mean()
std = time.std()

upper = mean + 1.5*std
lower = mean - 1.5*std

result = time[(time > upper) | (time < lower)].sum()
print(result)
```
